```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

```

```python
data=pd.read_csv("covid_19_india.csv")
df=pd.DataFrame(data)
df.drop(['Time','ConfirmedIndianNational','ConfirmedForeignNational','Sno'],axis=1,inplace=True)
print(df.head(10))
```

```python
data1=pd.read_csv("C:/Users/NITIN YADAV/Downloads/Covid-19-Data-Analysis-Project-Using-Python-And-Tableau-main/Covid-19-Data-Analysis-Project-Using-Python-And-Tableau-main/covid_vaccine_statewise.csv")
df1=pd.DataFrame(data1)
print(df1.head())
```

```python
df['Active cases']=df['Confirmed']-(df['Cured']+df['Deaths'])
df.tail(10)
```

```python
statewise=pd.pivot_table(df,index="State/UnionTerritory",values=["Confirmed","Deaths","Cured"],aggfunc="max")
statewise["Recovery rate"]=statewise["Cured"]*100/statewise["Confirmed"]
statewise["Mortality rate"]=statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
statewise=statewise.sort_values(by="Confirmed",ascending=False)
statewise.style.background_gradient(cmap="PuOr")
```

```python
top_ten=df.groupby(by="State/UnionTerritory").max()[['Active cases','Date']].sort_values(by=['Active cases'],ascending=False).reset_index()
print(top_ten)
```

```python

top_ten=df.groupby(by="State/UnionTerritory").max()[['Active cases','Date']].sort_values(by=['Active cases'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
Top_10_state_of_india=plt.title("Top 10 state of india with most active cases",size=25)
ax=sns.barplot(data=top_ten.iloc[:10],x="State/UnionTerritory",y="Active cases",edgecolor="red",hue="State/UnionTerritory")
plt.xlabel("states")
plt.ylabel("Active cases")
plt.show()
```

```python
# top 10 most deaths state 
top_ten=df.groupby(by="State/UnionTerritory").max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
Top_10_state_of_india=plt.title("Top 10 state of india with most Deaths cases",size=25)
ax=sns.barplot(data=top_ten.iloc[:12],x="State/UnionTerritory",y="Deaths",edgecolor="black",hue="State/UnionTerritory")
plt.xlabel("states")
plt.ylabel("Deaths")
plt.show()
```

```python
#growth trend in top 5 states
fig = plt.figure(figsize=(12, 6))
ax = sns.lineplot(data=df[df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerala', 'Uttar Pradesh', 'Tamil Nadu'])], x='Date', y='Active cases', hue='State/UnionTerritory')
ax.set_title("Top 5 affected States", size=25)
fig=plt.figure(figsize=(12,6))
```

```python
#remove rows where states='india'
vaccine=df1[df1.State!='India']
vaccine.head()
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated":"Total"},inplace=True)
vaccine.head(10)
```

```python
#most vaccinated state
max_vacc=vaccine.groupby('State')['Total'].sum().to_frame('Total').sort_values('Total',ascending=False)[:5]
max_vacc
```

```python
fig = plt.figure(figsize=(12, 6))
plt.title("Top 5 vaccination state in india",size=25)
x=sns.barplot(data=max_vacc.iloc[:5],x=max_vacc.index,y=max_vacc.Total,edgecolor="red",hue="State")
plt.xlabel("States")
plt.ylabel("vaccination")
plt.show()
```

```python
#least 5 vaccination states in india

least_vacc=vaccine.groupby('State')['Total'].sum().to_frame('Total').sort_values('Total')[:5]
least_vacc
```

```python
fig=plt.figure(figsize=(12,6))
x=sns.barplot(data=least_vacc.iloc[:5],x=least_vacc.index,y=least_vacc.Total,edgecolor="red",hue="State")
plt.xlabel("States")
plt.ylabel("vaccination")
plt.show()
```
